# Работа №2. Выборки и их представление

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Выборкой $x_1, ..., x_n$ объема $n$ из совокупности, распределенной по $F(x)$, называется $n$ независимых наблюдений над случайной величиной $\xi$ с функцией распределения $F(x)$.  
Вариационным рядом $x_{(1)} \le x_{(2)} \le ... \le x_{(n)}$ называется выборка, записанная в порядке возрастания ее элементов.  
Каждому наблюдению из выборки присвоим вероятность, равную ${1\over{n}}$; получим распределение, которое называют эмпирическим; ему соответствует функция эмпирического распределения
$$
F_n^*(x) = F_n^*(x; x_1, ..., x_n) = {\mu_n(x)\over{n}},
$$
где $\mu_n(x)$ - число членов выборки, меньших $x$. Значение этой функции для статистики определяется тем, что при $n \rightarrow \inf$
$$
F_n^*(x) \rightarrow F(x) \text{ теорема Гливенко.}
$$
Выборки больших объемов труднообозримы; разобьем диапазон значений выбрки на равные интервалы и подсчитаем для каждого интервала частоту - количество наблюдений, попавших в него; частоты, отнесенные к общему числу наблюдений $n$, называют относительными частотами; графическое представление распределения частот по интервалам - гистограммой; накопленной частотой для данного интервала называют сумму часотот данного интервала и всех тех, что левее его.  
Числовые характеристики эипирического распределения называются выборочными характеристиками: выборочные среднее(математическое ожидание), дисперсия:
$$
\overline{x} = {1\over{n}}\sum\limits_{i=1}^nx_i, \text{ } s^2 = {1\over{n}}\sum\limits_{i=1}^n(x_i-\overline{x})^2
$$
выборочный момент порядка k:
$$
m_k = {1\over{n}}\sum\limits_{i=1}^n(x_i-\overline{x})^k;
$$
выборочные квантики $\zeta_p$ порядка $p$ - корни уравнения
$$
F(\zeta_p) = p,
$$
которыми являются члены вариационного ряда
$$
\zeta_{(p)} = \xi_{[np]+1},
$$
где $[np]$ означает целую часть $np$; частным случаем ($p = 0.5$) является медиана - центральный член вариационного ряда. Значение выборочных характеристик состоит в том, что при $n \rightarrow \inf$ они стремятся к истинным значениям распределения $F(x)$.  
Приведем с помощью пакетов примеры. Исходные данные: R[0, 2], n = 50, a = 0.03

```python
n = 50
alpha = 0.03
data = np.array([np.random.uniform(0, 2) for _ in range(n)])
plt.plot(data)
plt.grid()
```

```python
data.sort()
for elem in data:
    print(elem)
```

```python
plt.hist(data, cumulative=True, bins=n)
plt.grid()
```

```python
# группирование данных
grouped = []
for i in range(10):
    grouped.append([])

b = 2
a = 0
step = (b-a)/10
i = 1
for elem in data:
    while elem - i*step > 0:
        i += 1
    grouped[i-1].append(elem)

for i in range(len(grouped)):
    print(f'Between {round(i*step, 2)} and {round((i+1)*step, 2)}:\n{grouped[i]}\nCount = {len(grouped[i])}\n\n')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
hist, xedges, yedges = np.histogram2d(x, y)

xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```
