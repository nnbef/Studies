### Сравнение оценок
#### Определения
Пусть $x_1, ..., x_n$ - выборка, то есть $n$ независимых испытаний случайной величины $X$, имеющей функцию распределения $F(x|a)$, зависящую от параметра $a$, значение которого неизвестно. Требуется оценить значение параметра $a$.  
Оценкой $a^{\text{^}} = \varphi (x_1, ..., x_n)$ называется функция наблюдений, используемая для приближенного определения неизвестного параметра. Значение $a^{\text{^}}$ оценки является случайной величиной, поскольку $(x_1, ..., x_n)$ - случайная величина (многомерная).  
**Свойства оценок**
1. Оценка $a^{\text{^}} = \varphi (x_1, ..., x_n)$ называется *состоятельной*, если при $n \rightarrow \inf$ $a^{\text{^}} \rightarrow a$ по вероятности при любом значении $a$.
2. Оценка $a^{\text{^}} = \varphi (x_1, ..., x_n)$ называется *несмещенной*, если при любом $a$ $Ma^{\text{^}} = M\varphi(x_1, ..., x_n) = a$  
Состоятельность -  обязательное свойство используемых оценок. Свойство несмещенности является желательным; многие применяемые оценки свойством несмещенности не обладают.
3. Оценка $\varphi^*$ называется *оптимальной*, если для нее средний квадрат ошибки $M(a^{\text{^}} - a)^2 = M[\varphi^*(x_1, ..., x_n) - a]^2 = \min M[\varphi(x_1, ..., x_n) - a]^2$ минимален среди всех оценок {$\varphi$}; здесь критерием качества оценки принят квадрат ошибки $(a^{\text{^}} - a)^2$. В более общей ситуации, если критерием качества служит некоторая величина $L(a^{\text{^}}, a)$, называемая функцией потерь (или функцией штрафа), то оптимальная оценк та, для которой минимальна величина $ML(a^{\text{^}}, a)$; последняя есть функция неизвестного $a$и называется функцией условного риста. Ясно, что оптимальной оценки может не существовать (так как характеристикой является функция, а не число).

#### Постановка конкретной задачи
**Пример.** Пусть на заводе имеется большая партия из $N$ (тысячи) транзисторов, используемых для сборки некоторого прибора. Выходные параметры прибора (например, надежность, уровень шума, вероятность выхода из режима и т.д.) зависят от обратных токов транзисторов; обратный ток у разных экземпляров различен, и потому можно считать его случайной величиной, причем, как известно технологам, распределённой равномерно в диапазоне от $0$ до $I_{max}$,  где $I_{max}$ — порог отбраковки, установленный на заводе - изготовителе транзисторов. Следовательно, выходные параметры прибора определяются величиной $I_{max}$. Предположим, что по каким-либо причинам значение $I_{max}$ производителю приборов неизвестно. Ясно, что в этом  случае из партии нужно случайным выбором извлечь $n$ (сравнительно немного: десятки) транзисторов, измерить их ток, и по измерениям оценить $I_{max}$ (неизвестный параметр $a$). Таким образом, возникает  
**Статистическая задача:** по наблюдениям $x_1, ..., x_n$ над случайной величиной $X$, распределенной равномерно на отрезке $[0, a]$, оценить неизвестный параметр $a$.  
Сравним три способа оценивания (три оценки):
1. оценку, полученную методом моментов
$$
a^{\text{^}}_1 = {2\over{n}}\sum\limits_{i=1}^nx_i;
$$
2. оценку, полученную методом максимального правдоподобия (после исправления смещенности)
$$
a^{\text{^}}_2 = {n+1\over{n}} \max x_i
$$
3. оценку, полученную методом порядковых статистик
$$
a^{\text{^}}_3 = 2*x^{\text{^}}_{0.5} = x_{(k)} + x_{(k+1)}, \text{ где } x^{\text{^}}_{0.5} = {x_{(k)} + x_{(k+1)}\over{2}} \text{ - выборочная квантиль порядка 0.5, то есть выборочная медиана, } x_{(k)} \text{ - член вариационного ряда с номером k, здесь полагаем n = 2k}
$$
Точность этих оценок можно сравнить теоретически и экспериментально (статистически).  
**Замечание.** Точность, однако, не является единственным критерием качества оценок. Весьма важно, например, свойство устойчивости оценки к изменению закона распределения или к засорению; в этом смысле, как оказывается, $a^{\text{^}}_3$ — наиболее хороша, а $a^{\text{^}}_2$ — наименее; действительно, пусть, например, в нашу выборку случайно попало наблюдение, резко превосходящее все остальные (в случае с партией триодов, попался триод, не прошедший отбраковку); значение оценки $a^{\text{^}}_2$ резко изменится, значение $a^{\text{^}}_3$ почти не изменится.

#### Теоретическое сравнение оценок
Все три оценки несмещенные, что можно проверить методами теории вероятностей. Определим дисперсии оценок:
$$
Da^{\text{^}}_1 = D({2\over{n}}\sum\limits_{i=1}^nx_i) = {a^2\over{3n}},\\
Da^{\text{^}}_2 = D({n+1\over{n}} \max x_i) = {a^2\over{n(n+2)}},\\
Da^{\text{^}}_3 = D(x_{(k)} + x_{(k+1)}) = {a^2\over{n}},
$$
откуда ясно, что $a^{\text{^}}_2$ - наиболее точоная оценка, а $a^{\text{^}}_3$ - наименее.

#### Статистическое сравнение оценок
Далеко не всегда удается аналитически вычислить дисперсию оценки. Как  экспериментально определить, какой из оценок пользоваться? По одной выборке нельзя судить о разбросе значений оценки, поскольку значение всего одно; необходимо иметь несколько выборок, например, $k = 20$, (или хотя бы 5 - 10), оценить разброс значений для каждой оценки и предпочесть ту оценку (тот способ оценивания), для которой разброс меньше. Если же выборка всего одна, то следует (если $n$ достаточно велико) разбить её случайным образом на несколько выборок, и по ним сравнивать качество оценок.  
Сформируем $k = 20$ выборок из распределения $R[0, a=10]$ объема $n$ для различных $n=10, 40, 160$ и определим разброс оценок. Характеристиками разброса значений $а_1,...,а_k$ оценки $a^{\text{^}}$ будем считать размах
$$
w = \max a_i - \min a_i
$$
и среднеквадратичное отклонение (СКО)
$$
S_a = \sqrt{{1\over{k-1}} \sum\limits_{i=1}^k (a_i - \overline{a})^2}, \text{ } \overline{a} = {1\over{k}} \sum\limits_{i=1}^k a_i
$$
Необходимо сравнить статистически на выборках объема $n=10, 40, 160$ две оценки: оценку максимального правдоподобия и медианную оценку.

```python
def generate_data(n):
    data = np.array([np.array([np.random.uniform(0, 10) for _ in range(n)]) for _ in range(20)])
    df = pd.DataFrame(data, columns=['']*n)
    a1, a2, a3 = [], [], []
    for i in range(len(data)):
        a1.append(2/n * data[i].sum())
        a2.append((n+1)/n * data[i].max())
        a3.append(2 * np.median(data[i]))
    a1 = np.array(a1)
    a2 = np.array(a2)
    a3 = np.array(a3)
    df['a1'] = a1
    df['a2'] = a2
    df['a3'] = a3
    return df
```

```python
n1, n2, n3 = 10, 40, 160
df1, df2, df3 = generate_data(n1), generate_data(n2), generate_data(n3)
df1
```

```python
pd.DataFrame([df1[['a1', 'a2', 'a3']].std(), df1[['a1', 'a2', 'a3']].min(), df1[['a1', 'a2', 'a3']].max()],
             index=['std', 'min', 'max'])
```

```python
df1[['a2', 'a3']].plot(title='Оценки a2 и a3 для n = 10', grid=True, ylim=(5, 15))
```

```python
df2[['a2', 'a3']].plot(title='Оценки a2 и a3 для n = 40', grid=True, ylim=(5, 15))
```

```python
df3[['a2', 'a3']].plot(title='Оценки a2 и a3 для n = 160', grid=True, ylim=(5, 15))
```

```python
pd.DataFrame(np.array([df1[['a1', 'a2', 'a3']].std(), df2[['a1', 'a2', 'a3']].std(), df3[['a1', 'a2', 'a3']].std()]),
             index=[10, 40, 160], columns=['a1', 'a2', 'a3']).plot(title='Сравнение оценок', grid=True)
```

#### Задача 1
Завод выпустил партию из $N = 1000$ приборов, предназначенных для непрерывной работы под водой в течение $T = 40$ часов. На испытания поставлены $n = 20$ приборов. Результаты испытаний на продолжительность оказались $x_1, x_2, ..., x_n$. Оценить количество неработоспособных приборов и их стоимость. Считать 1) продолжительность $X$ непрерывной работы случайной величиной, имеющей показательное распределение; 2) $C=200$ руб - стоимость одного прибора.  
Результаты получить моделированием, при моделировании принять среднее значение $m = 400$.

Данные распределены экспоненциально с параметром $a$ $X ~ E(a)$,
$$
p(x) = ae^{-ax}, x \ge 0
$$
Оценим параметр $a$ методом моментов:
$$
m_1 = \int\limits_{-\inf}^{\inf} xp(x)dx = \int\limits_{0}^{\inf} xae^{-ax}dx = {1\over{a}} \rightarrow a = {1\over{m_1}}
$$
Оценка для первого момента:
$$
m_1^{\text{^}} = {1\over{n}} \sum\limits_{i=1}^n\xi_i = \overline{\xi}\\
a^{\text{^}} = {1\over{\overline{\xi}}}
$$
Для того, чтобы оценить количество неработоспособных приборов, необходимо знать вероятность того, что прибор выйдет из строя:  
P - вероятность того, что прибор не работает
$$
P = P(\xi \le T) = \int\limits_{-\inf}^T p_X(x;a)dx = \int\limits_{0}^T ae^{-ax}dx = 1-e^{-aT} = P(a)
$$
Оценка для P:
$$
P^{\text{^}} = P(a^{\text{^}}) = 1 - e^{-a^{\text{^}}T}
$$
Выполним моделирование:

```python
N = 1000
n = 20
T = 40
C = 200
m = 400
data = np.array([np.random.exponential(m) for _ in range(n)])

print('Среднее:', data.mean())
```

Оценим a:
$$
a^{\text{^}} = {1\over{\overline{\xi}}} = 0.0024736627
$$
Оценим вероятность того, что прибор выйдет из строя:
$$
P^{\text{^}} = P(a^{\text{^}}) = 1 - e^{-a^{\text{^}}T} = 0.094
$$
Число неработоспособных приборов:
$$
k = N \cdot P = 1000 \cdot 0.094 = 94
$$
Стоимость неработоспособных приборов:
$$
S = C \cdot k = 200 \cdot 94 = 18800
$$
